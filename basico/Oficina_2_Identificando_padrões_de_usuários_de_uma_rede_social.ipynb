1. Encontrar usuários inativos: Precisamos encontrar os usuários que não fizeram login nos últimos 30 dias.
2. Enviar uma mensagem personalizada: Precisamos enviar uma mensagem para esses usuários, convidando-os a voltar a usar a rede social.

Aqui está um exemplo para ajudá-lo a começar:

1. Identificar usuários inativos:
  - Passo 1: Obter a lista de todos os usuários.
  - Passo 2: Para cada usuário, verificar a data do último login.
  - Passo 3: Se o último login foi há mais de 30 dias, adicionar o usuário à lista de inativos.

  - Passo 1: Para cada usuário inativo na lista, criar uma mensagem personalizada.
  - Passo 2: Enviar a mensagem para o usuário.

#### Pseudocodigo

```
INÍCIO

    lista_usuarios <- obter_lista_de_usuarios()  // Obtém a lista de todos os usuários da rede social para analisar quem está inativo
    lista_inativos <- []  // Cria uma lista vazia para armazenar os usuários inativos, facilitando a separação e o envio de mensagens depois

    data_atual <- DATA_ATUAL  // Armazena a data atual para calcular o período de inatividade
    limite_inatividade <- data_atual - 30 dias  // Define a data limite para considerar a inatividade (30 dias atrás), permitindo identificar usuários inativos

    // Percorre cada usuário na lista de usuários
    PARA CADA usuario EM lista_usuarios
        SE usuario.ultimo_login < limite_inatividade ENTÃO
            ADICIONAR usuario A lista_inativos  // Adiciona o usuário à lista de inativos se ele não acessa a rede há mais de 30 dias, indicando inatividade
        FIM-SE
    FIM-PARA

    // Envia mensagem para cada usuário inativo
    PARA CADA usuario INATIVO EM lista_inativos
        mensagem <- "Olá, " + usuario.nome + "! Sentimos sua falta na nossa rede social. Volte para conferir as novidades!"  // Cria uma mensagem personalizada para o usuário, incentivando-o a retornar
        EXIBIR "Enviando mensagem para " + usuario.nome + ": " + mensagem  // Exibe a mensagem simulando o envio, para registrar que o usuário foi notificado
    FIM-PARA

FIM

```
