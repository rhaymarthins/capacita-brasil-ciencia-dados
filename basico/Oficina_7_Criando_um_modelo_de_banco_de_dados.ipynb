Você e seus colegas são membros de uma equipe de consultoria de tecnologia contratada para ajudar uma empresa a melhorar a gestão dos seus dados. A empresa, que pode ser uma biblioteca, uma escola ou uma loja on-line, está enfrentando problemas com o gerenciamento e a organização dos dados que possui. Ela precisa de um novo sistema de banco de dados para armazenar e organizar suas informações de forma mais eficiente.

Sua missão é investigar e criar um modelo de banco de dados que atenda às necessidades específicas da empresa. Ela espera que você forneça uma solução bem-estruturada que possa ser implementada para resolver os problemas de gestão de dados dela.

Objetivo:

Investigar e criar um modelo de banco de dados que atenda às necessidades específicas da empresa.

Questão 1: Pesquise e explique por que os bancos de dados são importantes para a gestão de dados em empresas.

**Resposta:**

Bancos de dados são fundamentais para a gestão de dados nas empresas, pois possibilitam o armazenamento, organização e acesso rápido às informações. Eles ajudam a garantir que os dados sejam recuperados com precisão, facilitando a tomada de decisões e o controle das operações. Com uma estrutura bem planejada, um banco de dados também permite gerenciar grandes volumes de dados de maneira segura e eficiente, acompanhando o crescimento da empresa.

Questão 2: Pesquise as diferenças entre bancos de dados relacionais e NoSQL. Quais tipos de empresas se beneficiam mais de cada um?

**Resposta:**

Os bancos de dados relacionais organizam os dados em tabelas com linhas e colunas, sendo ideais para empresas que precisam de dados organizados e com relações complexas, como sistemas financeiros ou de inventário. Já os bancos de dados NoSQL são mais flexíveis, lidando bem com dados não estruturados, como textos ou imagens, sendo vantajosos para empresas de tecnologia, redes sociais ou e-commerce, que lidam com grandes volumes de dados em tempo real.
