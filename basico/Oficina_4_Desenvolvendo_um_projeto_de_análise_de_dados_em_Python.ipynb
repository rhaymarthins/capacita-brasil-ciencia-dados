Você está desenvolvendo um pequeno projeto de análise de dados em Python. Seu objetivo é processar uma lista de números, calcular a média dos valores, identificar quais números são maiores que a média e, finalmente, salvar os resultados em um arquivo CSV. Utilize seus conhecimentos sobre estruturas de dados, controle de fluxo, funções, e bibliotecas pandas e numpy para resolver esta questão.

Escreva um script Python que execute as seguintes etapas:


1. Use a lista de números: [12, 45, 67, 23, 89, 34, 22, 90, 56, 78].
2. Calcule a média dos números usando a biblioteca numpy.
3. Identifique os números que são maiores que a média.
4. Armazene os números maiores que a média em um DataFrame do pandas.
5. Salve o DataFrame em um arquivo CSV chamado "numeros_maiores_que_media.csv"

```python
import numpy as np
import pandas as pd

# Etapa 1
numeros = [12, 45, 67, 23, 89, 34, 22, 90, 56, 78]
numeros
```

```python
# Etapa 2
media = np.mean(numeros)
media
```

```python
# Etapa 3
maiores_que_media = [numero for numero in numeros if numero > media]
maiores_que_media
```

```python
# Etapa 4
df = pd.DataFrame(maiores_que_media, columns=["Numeros Maiores Que a Media"])
df
```

```python
# Etapa 5
df.to_csv('numeros_maiores_que_media.csv', index=False)
```
