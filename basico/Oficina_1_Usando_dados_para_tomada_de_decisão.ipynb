O diretor pediu sua ajuda porque percebeu que alguns alunos passam muito tempo esperando entre as aulas, enquanto outros têm dias mais corridos e cansativos.

Para resolver isso, você e seus colegas vão coletar informações sobre o tempo que cada um gasta para chegar à escola e o que passam nas aulas, nos intervalos e no deslocamento de volta para casa. A ideia é usar esses dados para entender como as aulas e os intervalos podem ser melhor distribuídos para tornar o horário escolar mais equilibrado para todos.

### 1. Coletar informações sobre o seu próprio dia escolar:
- Tempo gasto para chegar à escola (em minutos)
- Tempo de aula (por disciplina)
- Tempo de intervalo
- Tempo para deslocamento de volta para casa

### 2. Organizar esses dados em uma tabela simples;

```python
import pandas as pd

# Criando um DataFrame com dados fictícios
dados = {
    'Atividade': [
        'Deslocamento para escola',
        'Matemática',
        'Português',
        'Intervalo',
        'História',
        'Informática',
        'Deslocamento para casa'
    ],
    'Tempo (minutos)': [
        30, 50, 50, 30, 50, 50,  30
    ]
}

df = pd.DataFrame(dados)
df
```

### 3. Analisar quanto tempo você gasta em cada atividade do seu dia;

```python
# Tempo total gasto no dia escolar
tempo_total = df['Tempo (minutos)'].sum()
print(f'Tempo total gasto no dia escolar: {tempo_total} minutos')

# Porcentagem de tempo gasto em cada atividade
df['Porcentagem do tempo (%)'] = (df['Tempo (minutos)'] / tempo_total) * 100
df
```
