Imagine que você é o líder de um grupo de alunos da escola que está organizando um festival. O festival será sobre temas diversos, como música, filmes, esportes e muitos outros. Para atrair mais pessoas ao evento, vocês querem criar uma campanha com perguntas e respostas rápidas, que sejam interessantes e divertidas, para os participantes do festival. Porém, vocês precisam de ajuda para criar essas perguntas e encontrar informações rápidas sobre temas variados.

Você precisa gerar perguntas interessantes sobre diferentes temas para usar em um quiz interativo no festival. As perguntas devem ser simples, curiosas e podem cobrir temas variados como ciência, história, cultura pop, esportes e muito mais.

Usando um LLM (Large Language Model), um tipo de programa de inteligência artificial, você deve gerar 10 perguntas e respostas interessantes sobre temas variados para o quiz do festival. O foco é perceber como a IA pode ser usada para facilitar o processo de criação de conteúdo e para entender como ela pode responder a perguntas sobre temas do mundo real.

1. Definir temas: Escolha três temas que vocês gostariam de abordar no quiz (ex.: música, história, esportes).

2. Geração de perguntas: Usando o ChatGPT, Gemini ou qualquer outra plataforma de inteligência artificial (IA) generativa, gere 3 perguntas para cada tema escolhido. O objetivo é que as perguntas sejam interessantes e fáceis de entender para o público que irá participar do festival.

3. Verificar as respostas: Após gerar as perguntas, peça que a ferramenta forneça as respostas para cada uma delas. Verifique se as respostas fazem sentido e estão corretas.

4. Escolher as melhores perguntas: Depois de gerar as perguntas e respostas, escolham as 10 melhores que serão usadas no quiz do festival e envie.

Perguntas e Respostas geradas para cada tema.

**Tema: Música**

  1. Pergunta: Qual é o nome da icônica música do Queen que mistura ópera, rock e balada, durando mais de 6 minutos?

  Resposta: "Bohemian Rhapsody".

  2. Pergunta: Qual cantora pop é conhecida como a "Rainha do Pop" e tem um álbum chamado "Like a Virgin"?

  Resposta: Madonna.

  3. Pergunta: Em que ano Michael Jackson apresentou pela primeira vez o "moonwalk" ao mundo, durante um show da Motown?

  Resposta: 1983 (no especial Motown 25: Yesterday, Today, Forever).

**Tema: Games/Geek**

  4. Pergunta: Qual é o nome do encanador bigodudo criado pela Nintendo, que salva princesas em um reino de cogumelos?

  Resposta: Mario (originalmente chamado de Jumpman no jogo Donkey Kong).

  5. Pergunta: Qual jogo de terror indie, lançado em 2014, popularizou a mecânica de esconder de animatrônicos em uma pizzaria assombrada?

  Resposta: Five Nights at Freddy's (o primeiro jogo da franquia).

  6. Pergunta: Em qual jogo os jogadores constroem e exploram mundos com blocos, enfrentando criaturas como Creeper?

  Resposta: Minecraft.

**Tema: Filmes**

  7. Pergunta: Qual filme de 1997, dirigido por James Cameron, ganhou 11 Oscars e tem a famosa cena do "Estou voando, Jack!"?
  
  Resposta: Titanic.

  8. Pergunta: Qual filme da Disney de 2013 viralizou com a música 'Let It Go'?
  
  Resposta: Frozen - Uma Aventura Congelante.

  9. Pergunta: Qual foi o primeiro filme da história a ultrapassar a marca de US$ 2 bilhões em bilheteria mundial?
  
  Resposta: Avatar (2009), dirigido por James Cameron.

  10. Pergunta: Qual ator interpretou o Coringa no filme Coringa (2019), ganhando um Oscar por sua atuação?
  
  Resposta: Joaquin Phoenix.
