````python
# Passo 1: Gerando o Dataset Sintético
# Importando as bibliotecas necessárias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Gerando o dataset sintético
np.random.seed(0)
n_samples = 300

data = {
    'valor_medio_gasto': np.random.randn(n_samples) * 20 + 200,
    'frequencia_compras': np.random.randn(n_samples) * 5 + 15,
    'categorias_produtos': np.random.randn(n_samples) * 2 + 8
}

df = pd.DataFrame(data)
df.head()
````

````python
# Passo 2: Aplicando o Algoritmo K-Means
# Selecionando as features para o clustering
X = df[['valor_medio_gasto', 'frequencia_compras', 'categorias_produtos']].values

# Definindo o número de clusters (k=3)
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)

# Prevendo os clusters
df['cluster'] = kmeans.predict(X)
````

````python
# Passo 3: Visualizando os Resultados
# Visualizando os clusters para 'valor_medio_gasto' e 'frequencia_compras'
plt.scatter(df['valor_medio_gasto'], df['frequencia_compras'], c=df['cluster'], s=50, cmap='viridis')

# Plotando os centróides
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')

plt.title("Clustering com K-means")
plt.xlabel("Valor Médio Gasto")
plt.ylabel("Frequência de Compras")
plt.show()
````

### Justificativa da Escolha do Algoritmo

O K-Means é a escolha mais adequada para este problema de segmentação de clientes, pois estamos lidando com dados numéricos e buscamos agrupar os clientes com base em características como valor médio gasto, frequência de compras e número de categorias de produtos comprados. O K-Means é eficiente para identificar padrões em dados não rotulados, sendo ideal para essa tarefa.

O K-Means possui vantagens significativas, como simplicidade e facilidade de implementação, eficiência em grandes conjuntos de dados e boa interpretabilidade dos clusters formados. No entanto, ele também tem limitações: requer a definição prévia do número de clusters (k), o que pode ser desafiador sem um entendimento claro dos dados, e é sensível a outliers e à inicialização dos centróides.

Os clusters formados podem ser interpretados da seguinte forma:
- Cluster 0: Representa clientes que gastam menos, compram com menor frequência e em menos categorias.
- Cluster 1: Corresponde a clientes com gastos moderados, frequência média de compras e um número moderado de categorias.
- Cluster 2: Reflete clientes que gastam mais, compram com maior frequência e em mais categorias.

Esses insights permitem que a loja on-line direcione campanhas de marketing específicas, ofereça descontos personalizados e ajuste o estoque conforme as preferências de cada grupo, otimizando a experiência do cliente e aumentando a eficiência das estratégias de negócio.
