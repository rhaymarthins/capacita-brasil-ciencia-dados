1. De que maneira o uso do Word2Vec ajudaria o chatbot a entender diferentes variações das perguntas dos usuários?

  **Resposta:**

  O Word2Vec converte palavras em vetores numéricos, identificando relações entre termos. Por exemplo:
  - Se o usuário pergunta "Como rastrear meu pedido?" ou "Onde está minha encomenda?", o Word2Vec entende que "rastrear" e "onde está" estão associados ao conceito de localização de um pedido.
  - Isso permite que o chatbot reconheça diferentes formas de perguntar a mesma coisa, respondendo de maneira precisa mesmo com palavras ou estruturas frasais distintas.

2. Como o uso de redes neurais recorrentes pode impactar a continuidade do diálogo em uma conversa mais longa com o chatbot? Para ajudá-lo a guiar na sua resposta veja alguns aspectos que você pode levar em consideração:
  - Apresentar um plano claro de como utilizar as tecnologias mencionadas para otimizar o chatbot;
  - Explicar brevemente o papel de cada técnica no aprimoramento do chatbot;
  - Demonstrar compreensão de como integrar essas ferramentas para criar uma interação mais eficiente com os usuários.

  **Resposta:**

  As redes neurais recorrentes (especialmente variantes como LSTM ou GRU) processam sequências de texto mantendo um "memória" do contexto anterior. Isso é essencial para conversas longas:

  Exemplo: Se o usuário diz "Quero cancelar minha assinatura" e depois pergunta "Como faço isso?", a RNN entende que "isso" se refere ao cancelamento.
  - Plano de integração das técnicas:
    - Word2Vec: Transforma palavras como "defeito", "trocar" e "devolução" em vetores similares, identificando que todas estão ligadas a problemas com produtos.
    - RNN/LSTM: Mantém o histórico da conversa. Exemplo: Se o usuário mencionou "Meu produto chegou quebrado", a RNN lembra desse detalhe ao responder perguntas subsequentes, como "Posso receber um reembolso?".
    - Análise de Sentimento: Detecta emoções no texto. Exemplo: Se o usuário escreve "Estou extremamente insatisfeito!", o chatbot prioriza respostas rápidas e oferece ajuda humana.
  - Papel de cada técnica:
    - Usuário: "Quero devolver um item, como funciona?"
      - Word2Vec: Associa "devolver" a termos como "reembolso" e "troca".
      - RNN: Armazena que o assunto é "devolução".
    - Usuário (na próxima mensagem): "Preciso pagar alguma taxa?"
      - RNN: Reconhece que a pergunta está relacionada à devolução anterior.
      - Análise de Sentimento: Detecta neutralidade → Resposta direta: "Não há taxas para devoluções em até 7 dias."
  - Integração de cada técnica:
    - Word2Vec → "O que o usuário quer?".
    - RNN → "Do que estamos falando?".
    - Análise de Sentimento → "Como devo responder?".
