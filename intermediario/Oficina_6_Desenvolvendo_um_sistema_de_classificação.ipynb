# 1. Contextualização

O sistema de e-mail da empresa não está funcionando como deveria. Muitos e-mails indesejados (spam) estão passando pelo filtro atual, o que atrapalha bastante os funcionários. A ideia é criar um sistema que resolva esse problema, separando com mais precisão os e-mails "spam" dos "não spam". Felizmente, temos um banco de dados com e-mails antigos já classificados, o que nos dá uma boa base para treinar o modelo.

# 2. Escolha do Algoritmo

O algoritmo escolhido é a Árvore de Decisão.

# 3. Justificativa Técnica

A árvore de decisão é bem intuitiva. Imagine ela como um conjunto de perguntas que levam a um resultado. Para classificar um e-mail, a árvore vai perguntar coisas como: "O e-mail contém muitas palavras relacionadas a promoções?", "O remetente já foi marcado como spam antes?", e assim por diante. Com base nas respostas, ela vai percorrendo as ramificações até chegar à decisão final: spam ou não spam.

# 4. Adequação ao Problema

A árvore de decisão é super útil para esse tipo de problema porque ela transforma dados em regras fáceis de entender, como "Se o e-mail contém 'promoção' e vem de um remetente desconhecido, é spam". Isso é ótimo quando precisamos explicar para os colegas ou gestores como o sistema está tomando decisões. Além disso, ela é rápida tanto para treinar quanto para classificar e-mails.

Mas tem uma questão: se não ajustar bem o modelo, ele pode acabar se adaptando demais aos dados antigos, causando overfitting, e ter dificuldade em reconhecer padrões em e-mails novos. Para evitar isso, dá para limitar a profundidade da árvore ou garantir que cada folha tenha uma quantidade mínima de exemplos antes de ser criada.

# 5. Considerações Adicionais

Um ponto importante é decidir bem quais características do e-mail vamos usar para treinar a árvore, como as palavras mais comuns no texto, se o remetente é conhecido ou se o e-mail contém links ou anexos. Isso faz toda a diferença no desempenho do modelo.

Também vale lembrar que, conforme o número de e-mails no banco de dados aumenta, treinar uma nova árvore pode levar um pouco mais de tempo, mas o sistema continuará sendo rápido para classificar e-mails na prática. Se quisermos melhorar ainda mais o modelo, dá para usar técnicas como "poda" (pruning) ou até pensar em criar várias árvores (como no método Random Forest) para combinar resultados e ter um filtro ainda mais preciso.
