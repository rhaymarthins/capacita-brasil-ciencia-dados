Agora, sua tarefa é preparar esse conjunto de dados para construir um modelo preditivo que ajudará a empresa a segmentar melhor seus clientes em campanhas futuras.

# 1. Limpeza de Dados e Imputação de Dados
- **Coluna Idade:**

Para os dados ausentes imputaria com a mediana porque ela representa melhor o valor típico da idade, ignorando valores muito altos ou muito baixos que poderiam distorcer a média.

- **Coluna Gênero:**

Para os dados ausentes imputaria com o valor que mais aparece (moda) porque em categorias como gênero, é mais lógico usar um valor que já existe nos dados do que criar um novo.

- **Coluna Renda Anual (R$):**

Para os dados ausentes imputaria com a média porque, em dados de renda, ela nos dá uma boa ideia do valor típico considerando todas as rendas registradas.

- **Compras Anteriores:**

Para os dados ausentes imputaria com  a mediana porque alguns clientes podem ter feito muitas ou poucas compras, e a mediana nos dá um número mais realista do padrão típico de compras.

# 2. Transformação de Dados

- **Renda Anual:**

Aplicaria normalização transformaando os dados para escala 0-1 para tornar mais fácil comparar rendas muito diferentes. Por exemplo, R\$ 30.000 vira 0,0 e R\$100.000 vira 1,0.

- **Idade:**

Aplicaria padronização ajustando os valores para que fiquem mais equilibrados entre si, evitando que idades maiores (como 50) tenham mais peso que idades menores (como 23) no modelo.

- **Compras Anteriores:**

Aplicaria padronização ajustando os números de compras da mesma forma que a idade, para que clientes com muitas compras (como 8) não tenham influência exagerada comparado aos que fizeram poucas compras (como 1).


# 3. Codificaçao de Dados

- **Resposta à Campanha:**

Tranformaria a coluna em "Sim" em 1 e "Não" em 0, porque computadores trabalham melhor com números do que com texto.

- **Localização e Gênero:**

Tranformaria colunas separadas para cada cidade e gênero, marcando 1 onde o cliente pertence e 0 onde não pertence, porque não faz sentido dizer que uma cidade é "maior" que outra.
