Você foi recentemente contratado como analista de dados em uma empresa de tecnologia que desenvolve soluções para a gestão de operações em empresas de diversos setores. Parte do seu trabalho envolve a análise de relatórios sobre incidentes que ocorreram em diferentes instalações da empresa.

Esses relatórios são importantes, pois ajudam os gestores a identificar padrões, entender as causas dos incidentes e tomar decisões para prevenir problemas futuros.

O relatório atual apresenta gráficos que incluem dados, como status dos incidentes, número de incidentes por instalação, categorias de incidentes e causas contribuintes. Ao examinar os gráficos e relatórios fornecidos, você se deparou com estas informações:

unnamed.jpg

Analisando a imagem, responda às seguintes questões:


a) Quais são os principais problemas que você identificou nas visualizações fornecidas?

**Resposta:**

- Sobrecarregamento de informações: O dashboard apresenta muitas informações simultaneamente, o que pode dificultar a concentração nos dados mais relevantes.
- Uso inconsistente de cores: Diferentes gráficos usam esquemas de cores distintos para representar categorias similares, o que pode causar confusão.
- Legendas truncadas: Algumas descrições nas legendas estão cortadas.
- Escala dos gráficos: Alguns gráficos têm escalas que dificultam a interpretação e análise dos dados.
- Falta de hierarquia visual: Não há clara distinção entre informações primárias e secundárias.

b) Como esses problemas podem impactar a interpretação dos dados pelos tomadores de decisão?

**Resposta:**

- Sobrecarga cognitiva: O excesso de informações pode causar fadiga mental, dificultando a tomada de decisões e a identificação de padrões relevantes.
- Interpretações errôneas: A inconsistência no uso de cores pode levar a conclusões equivocadas sobre relações entre diferentes métricas.
- Perda de contexto: Com legendas truncadas, os tomadores de decisão podem perder informações cruciais sobre as causas dos incidentes.
- Dificuldade em priorização: A falta de hierarquia visual clara pode dificultar a identificação das áreas que precisam de atenção imediata.

c) Para mostrar a comparação de desempenho entre diferentes tipos de incidentes, qual tipo de gráfico você recomendaria e por quê?

**Resposta:**

Recomendaria um gráfico de barras horizontais com as seguintes características:

- Ordenação por frequência/severidade.
- Código de cores consistente.
- Possibilidade de agrupamento por categorias.
- Rótulos claros e completos.

d) Se você precisasse exibir a evolução dos incidentes ao longo do tempo e o status dela para diferentes grupos demográficos, qual seria o tipo de gráfico mais apropriado?

**Resposta:**

Recomendaria um gráfico de área empilhada com as seguintes características:

- Eixo X mostrando a linha do tempo.
- Áreas empilhadas mostrando diferentes grupos.
- Cores distintas para cada status.
- Possibilidade de interação para detalhar períodos específicos.
- Legendas claras e completas.

Justificativa:

- Permite visualizar tendências ao longo do tempo.
- Mostra a contribuição relativa de cada grupo.
- Facilita a identificação de padrões sazonais.
- Permite comparação entre diferentes períodos.
- Mantém o contexto total enquanto mostra as partes individuais.
